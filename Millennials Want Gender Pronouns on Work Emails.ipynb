```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Data Preparation
# Pronoun usage awareness
pronoun_data = {
    'Generation': ['Gen Z', 'Millennials', 'Gen X', 'Boomers'],
    'Always Use Pronouns': [60, 45, 30, 15],
    'Occasionally Use Pronouns': [25, 35, 40, 30],
    'Not Aware of Usage': [15, 20, 30, 55]
}

pronoun_df = pd.DataFrame(pronoun_data)

# Email formality preference
email_data = {
    'Preference': ['Formal Emails', 'Informal Emails', 'Neutral'],
    'Gen Z': [40, 50, 10],
    'Millennials': [45, 40, 15],
    'Gen X': [60, 30, 10],
    'Boomers': [70, 20, 10]
}

email_df = pd.DataFrame(email_data)

# Visualization 1: Stacked Bar Chart for Pronoun Usage
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.6
bar_positions = np.arange(len(pronoun_df['Generation']))

# Stacked bar components
colors = ['#0072B2', '#F0E442', '#006A4E']
ax.bar(bar_positions, pronoun_df['Always Use Pronouns'], bar_width, label='Always Use Pronouns', color=colors[0], edgecolor='black')
ax.bar(bar_positions, pronoun_df['Occasionally Use Pronouns'], bar_width, bottom=pronoun_df['Always Use Pronouns'], label='Occasionally Use Pronouns', color=colors[1], edgecolor='black')
ax.bar(bar_positions, pronoun_df['Not Aware of Usage'], bar_width, bottom=(pronoun_df['Always Use Pronouns'] + pronoun_df['Occasionally Use Pronouns']), label='Not Aware of Usage', color=colors[2], edgecolor='black')

# Adding annotations
for i in range(len(pronoun_df)):
    total = sum(pronoun_df.iloc[i, 1:])
    ax.text(bar_positions[i], total + 2, f"{total}%", ha='center', fontsize=10, color='black', fontweight='bold')

ax.set_xticks(bar_positions)
ax.set_xticklabels(pronoun_df['Generation'], fontsize=12)
ax.set_ylabel('Percentage', fontsize=12)
ax.set_xlabel('Generation', fontsize=12)
ax.set_title('Pronoun Usage Awareness Across Generations', fontsize=16, fontweight='bold', color='#333333')
ax.legend(title='Pronoun Usage', fontsize=10, title_fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0., color='gray')
plt.tight_layout()
plt.show()

# Visualization 2: Pie Chart for Email Formality Preference
gen_z_prefs = email_df[['Preference', 'Gen Z']]
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(gen_z_prefs['Gen Z'], labels=gen_z_prefs['Preference'], autopct='%1.1f%%', startangle=140, colors=['#8DA0CB', '#FC8D62', '#66C2A5'],
                                   textprops={'fontsize': 12, 'color': '#333333'}, pctdistance=0.85)

# Adding a circle for a donut chart effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Gen Z Preference for Email Formality', fontsize=16, fontweight='bold', color='#333333')
plt.tight_layout()
plt.show()



```
